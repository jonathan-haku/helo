## Using DrawingPanel to Create Graphics

The **DrawingPanel.py** file provides a convinient package to draw simple graphic shapes. It is based on the Python tkinter package, which allows Python program to use Tk/Tcl to create Graphical User Interface (GUI) and graphics applications.

DrawingPanel package defines a panel object that has functions to draw shapes. 

To use this package, you must 

- place the DrawingPanel.py file in the directory where your own program will be, and 

- import the package within your program

      from DrawingPanel import *

- To create a panel, you do the following

      panel = DrawingPanel(width, height)
     
- If the panel is not shown, enter the following statement

      panel.window.mainloop()
     
A window with the panel will be displayed at the center of the background. You may have to move other windows away to see the panel window.

## Coordinate System
- The window's drawing area consists of $width\times height$ pixels (picture element) and each pixel is at a position ($x$, $y$)

- The position (0, 0) is at the upper-left corner of the window.

  - x increases towards the right
  - y increases towards the bottom

## Colors

- Colors can be specified in three ways.

    - as a name string, such "red" and "blue" 
    - as a color constant, such as, Color.RED or Color.GREEN
    - as a color object defined by a RGB (stands for Red, Gree, and Blue) trio, such as (255, 0, 182), where each of the three numbers can be int between 0 and 255.


- A complete list of color names can be found at http://www.tcl.tk/man/tcl8.5/TkCmd/colors.htm

## Mouse Event Handler

- The system will catch any mouse movement and the pixel position of the mouse.

- Add a mousr_listener to the panel, and pass a click_handler function as an argument.

         panel.add_mouse_listener(click_handler)

         def click_handler(click_event) :
             ......
                    
- This function will be called by the run-time system each time the mouse is clicked on the panel. The click_event has two attributes x and y, which provide the coordinates on the panel where the mouse was clicked.

## Example

Write a function to capture mouse click position and determine if it is in the center of the panel.


```python
from DrawingPanel import *

def click_handler(click_event):
    x = click_event.x
    y = click_event.y
    print("You clicked at position (%d, %d)." % (x, y))
    if 150 <= x <= 250 and 150 <= y <=250 :
        print("This is inside the center area.")
    else :
        print("This is outside of the center area.")
    
def main():
    WIDTH = 401
    HEIGHT = 401
    panel = DrawingPanel(WIDTH, HEIGHT)
    panel.add_mouse_listener(click_handler)
    panel.window.mainloop()

if __name__ == "__main__":
    main()
```
